```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    """The function f(x) = x^4 + 2x^3 - 5x^2 + 40."""
    return x**4 + 2*x**3 - 5*x**2 + 40

def df(x):
    """The derivative of f(x)."""
    return 4*x**3 + 6*x**2 - 10*x

def ddf(x):
    """The second derivative of f(x)."""
    return 12*x**2 + 12*x - 10

# Find critical points numerically (using numpy's roots function)
# We solve df(x) = 0
coefficients = [4, 6, -10, 0] # Coefficients of the derivative
critical_points = np.roots(coefficients)
print("Critical Points:", critical_points)

# Evaluate the second derivative at the critical points
for cp in critical_points:
    second_derivative_value = ddf(cp)
    print(f"f''({cp:.2f}) = {second_derivative_value:.2f}", end=" ")
    if second_derivative_value > 0:
        print("(Local Minimum)")
    elif second_derivative_value < 0:
        print("(Local Maximum)")
    else:
        print("(Inflection Point or inconclusive)")

# Find the minimum value among the local minima
min_x = None
min_f_x = float('inf')  # Initialize with positive infinity

for cp in critical_points:
    if ddf(cp) > 0: #Check if it is a local minimum
        if f(cp) < min_f_x:
            min_f_x = f(cp)
            min_x = cp

if min_x is not None:
    print(f"\nThe global minimum occurs at x = {min_x:.2f} with f(x) = {min_f_x:.2f}")
else:
    print("No minimum found")

# Plot the function and its derivatives (optional visualization)
x_vals = np.linspace(-4, 2, 400)  # Create x values for plotting
plt.plot(x_vals, f(x_vals), label="f(x)")
plt.plot(x_vals, df(x_vals), label="f'(x)")
plt.plot(x_vals, ddf(x_vals), label="f''(x)")
plt.scatter(critical_points, f(critical_points), color='red', label="Critical Points") #Mark Critical Points
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x) and its Derivatives")
plt.legend()
plt.grid(True)
plt.show()
```
