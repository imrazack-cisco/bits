```python
import numpy as np
import matplotlib.pyplot as plt

# Set up X and y
X = np.array([[1, 3], [1, 3], [1, 6], [1, 1], [1, 8], [1, 9]])
y = np.array([30, 36, 43, 20, 57, 64])

def normal_equation(X, y):
    return np.linalg.inv(X.T @ X) @ X.T @ y

# Calculate w using the normal equation
w = normal_equation(X, y)

print("Regression coefficients:", w)

# Predict for 10 years of experience
x_new = np.array([1, 10])
y_pred = x_new @ w

print("Predicted salary for 10 years of experience:", y_pred, "lakhs")

# Plot the data and regression line
plt.scatter(X[:, 1], y, label="Data")
plt.plot(X[:, 1], X @ w, color='red', label="Regression Line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary (Lakhs)")
plt.title("Salary Prediction")
plt.legend()
plt.show()
```
