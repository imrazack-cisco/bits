```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Prior distribution (Beta(1, 1))
a_prior = 1
b_prior = 1

# Observed data
heads = 7
tails = 3

# Posterior distribution (Beta(a_prior + heads, b_prior + tails))
a_posterior = a_prior + heads
b_posterior = b_prior + tails

# Calculate the MAP estimate
map_estimate = (a_posterior - 1) / (a_posterior + b_posterior - 2)

print("MAP estimate of the probability of heads:", map_estimate)

# Visualize the prior and posterior distributions
x = np.linspace(0, 1, 100)
prior_pdf = beta.pdf(x, a_prior, b_prior)
posterior_pdf = beta.pdf(x, a_posterior, b_posterior)

plt.plot(x, prior_pdf, label='Prior')
plt.plot(x, posterior_pdf, label='Posterior')
plt.xlabel('Probability of Heads (θ)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
