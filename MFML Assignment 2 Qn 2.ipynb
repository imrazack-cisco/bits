```python
import matplotlib.pyplot as plt
import math

def f(x):
    """The function f(x) = x^4 + 2x^3 - 5x^2 + 40."""
    return x**4 + 2*x**3 - 5*x**2 + 40

def df(x):
    """The derivative (gradient) of f(x)."""
    return 4*x**3 + 6*x**2 - 10*x

def gradient_descent(x0, alpha, max_iterations=11, tolerance=0.0001):
    """Performs gradient descent."""
    x = x0
    iterations = []
    gradient_magnitudes = []

    print(f"alpha = {alpha}")
    print("--------------------------------")
    print("Iteration |       x      |   |∇f(x)|  ")
    print("--------------------------------")

    for i in range(max_iterations):
        gradient = df(x)
        gradient_magnitude = abs(gradient)
        iterations.append(i)
        gradient_magnitudes.append(gradient_magnitude)

        print(f"{i+1:^9} | {x:^12.4f} | {gradient_magnitude:^10.4f}")

        if gradient_magnitude < tolerance:
            print("--------------------------------")
            print("Tolerance reached")
            print("--------------------------------")
            break

        x = x - alpha * gradient
    else:
        print("--------------------------------")
        print("Max iterations reached")
        print("--------------------------------")

    return iterations, gradient_magnitudes, x

# Initial value
x0 = -1

# Learning rates
alphas = [0.01, 0.1]

for alpha in alphas:
    iterations, gradient_magnitudes, final_x = gradient_descent(x0, alpha)

    # Plot
    plt.plot(iterations, gradient_magnitudes, marker='o', label=f"alpha = {alpha}")
    print(f"Final x value for alpha = {alpha}: {final_x:.4f}")
    print()

plt.xlabel("Iteration Number")
plt.ylabel("|∇f(x)|")
plt.title("Gradient Descent Convergence")
plt.legend()
plt.grid(True)
plt.yscale('log') #Use log scale for y to better visualize convergence.
plt.show()

#Print final f(x) values
for alpha in alphas:
    iterations, gradient_magnitudes, final_x = gradient_descent(x0, alpha)
    final_fx = f(final_x)
    print(f"Final f(x) value for alpha = {alpha}: {final_fx:.4f}")
    print()
```
